```python
import datetime
import json
try:
    import modsim
except ImportError:
    !pip install modsimpy
    !pip install pint   
    import modsim

Lista_de_ID_Doentes = []
###########################################################################################################################
def MenuPrincipal():
    print('\n--------MENU PRINCIPAL--------')
    print('\n1 – Introduzir novo doente')
    print('\n2 – Modificar sintomas')
    print('\n3 – Testar doente')
    print('\n4 – Remover doente')
    print('\n5 – Consultar sistema')
    print('\n6 – Gravar dados')
    print('\n7 - Carregar dados')
    print('\n8 – Simulação de ocupação UCI')
    print('\n0 – Sair')
    return ()
###########################################################################################################################
def AdicionarDoente():
    doente = {}
    print('\n--------------------ADICIONAR UM NOVO DOENTE--------------------')
    nome = input('\nInsira o nome: ')
    while nome.isalpha() == False:
        nome = input("\nInsira o nome: ")
    doente['nome']=nome  
    numero_telefone = str(input('\nInsira o nº telefone: '))
    while len(str(numero_telefone)) != 9 or not numero_telefone.isdigit():
        numero_telefone = input('\nInsira o nº telefone (9 dígitos): ')
    doente['numero_telefone'] = numero_telefone
    morada = str(input('\nInsira a morada: '))
    doente['morada'] = morada
    cc = input(str('\nInsira o número do cartão de cidadão: '))
    while len(str(cc)) != 8 or not cc.isdigit():
        cc = input('\nInsira o número do cartão de cidadão (8 dígitos): ')
    (doente['cc']) = cc
    print('\nInsira a data de nascimento:')
    dia = str(input('Dia: '))
    while not dia.isdigit() or int(dia) > 31 or int(dia) <= 0 :
        dia = input('Dia: ')
    mes = str(input('Mes: '))
    while not mes.isdigit() or int(mes) > 12 or int(mes) <= 0  :
        mes = input('Mes: ')
    ano = str(input('Ano: '))
    while not (ano.isdigit()) or int(ano)<1000 or int(ano)>9999:
        ano = str(input('Ano: ')) 
    dia = int(dia)
    mes = int(mes)
    ano = int(ano)
    doente['data_nasc'] = str(dia) + '-' + str(mes) + '-' + str(ano)
    tosse = str(input('\nTosse? (s/n): '))
    while (tosse != 's') and (tosse != 'n'):
        tosse = str(input('\nTosse? (s/n): '))
    (doente['tosse'])=(tosse)
    temp = str(input('\nTemperatura superior a 37.8? (s/n): '))
    while (temp != 's') and (temp != 'n'):
        temp = str(input('\nTemperatura superior a 37.8? (s/n): '))
    (doente['temp'])=(temp)
    dores_m = str(input('\nDores musculares? (s/n): '))
    while (dores_m != 's') and (dores_m != 'n'):
        dores_m = str(input('\nDores musculares? (s/n): '))
    (doente['dores_m'])=(dores_m)
    cef = str(input('\nCefaleias? (s/n): '))
    while (cef != 's') and (cef != 'n'):
        cef = str(input('\nCefaleias? (s/n): '))
    (doente['dores_c'])=(cef)
    fraqueza = str(input('\nFraqueza? (s/n): '))
    while (fraqueza != 's') and (fraqueza != 'n'):
        fraqueza = str(input('\nFraqueza? (s/n): '))
    (doente['fraqueza'])=(fraqueza)
    dif_resp = str(input('\nDificuldade respiratória? (s/n): '))
    while (dif_resp != 's') and (dif_resp != 'n'):
        dif_resp = str(input('\nDificuldade respiratória? (s/n): '))
    (doente['dif_resp'])=(dif_resp)
    (doente['percentagem'])=('Por Avaliar')
    (doente['risco'])=('Por Avaliar')
    Lista_de_ID_Doentes.append(doente)
    add = str(input('\nDeseja adicionar outro doente? (s/n): '))
    while (add != 's') and (add != 'n'):
        add = str(input('\nDeseja adicionar outro doente? (s/n): '))
    if add == 's':
        print('\nDoente adicionado com sucesso!')
        AdicionarDoente()
    else:
        print('\nDoente adicionado com sucesso!')
        return()
###########################################################################################################################
def ModificarSintomas():
    
    print('\n--------------------MODIFICAR SINTOMAS--------------------')
    nome = str(input('\nInsira o nome: '))
    c=len(Lista_de_ID_Doentes)
    for d in Lista_de_ID_Doentes:
        if d["nome"]==nome:
            tosse = str(input('\nTosse? (s/n): '))
            while (tosse != 's') and (tosse != 'n'):
                tosse = str(input('\nTosse? (s/n): '))
            d["tosse"] = tosse
            temp = str(input('\nTemperatura superior a 37.8? (s/n): '))
            while (temp != 's') and (temp != 'n'):
                temp = str(input('\nTemperatura superior a 37.8? (s/n): '))
            d['temp']=temp
            dores_m = str(input('\nDores musculares? (s/n): '))
            while (dores_m != 's') and (dores_m != 'n'):
                dores_m = str(input('\nDores musculares? (s/n): '))
            (d['dores_m'])=(dores_m)
            cef = str(input('\nCefaleias? (s/n): '))
            while (cef != 's') and (cef != 'n'):
                cef = str(input('\nCefaleias? (s/n): '))
            (d['dores_c'])=(cef)
            fraqueza = str(input('\nFraqueza? (s/n): '))
            while (fraqueza != 's') and (fraqueza != 'n'):
                fraqueza = str(input('\nFraqueza? (s/n): '))
            (d['fraqueza'])=(fraqueza)
            dif_resp = str(input('\nDificuldade respiratória? (s/n): '))
            while (dif_resp != 's') and (dif_resp != 'n'):
                dif_resp = str(input('\nDificuldade respiratória? (s/n): '))
            (d['dif_resp'])=(dif_resp)
            print('Dados modificados com sucesso!')
        else:
            c-=1
        if c == 0:
            print("Erro - doente não existe. Crie primeiro uma entrada…")
        md = str(input('Deseja modificar outro doente? (s/n): '))
        while (md != 's') and (md != 'n'):
            md = str(input('Deseja modificar outro doente? (s/n): '))
        if md == 's':
            ModificarSintomas()
        else:
            return()
###########################################################################################################################
def RemoverDoente():
    print('\n--------------------REMOVER DOENTE--------------------')
    nome = str(input('\nInsira o nome: '))
    c=len(Lista_de_ID_Doentes)
    
    for d in Lista_de_ID_Doentes:
        if d['nome']== nome:
            print("Para remover um doente este têm de ter ""Alta Médica"" ou ""Transferido"".")
            motivo = str(input('\nInsira o motivo: '))
            while motivo != ('Alta Médica') and motivo != ('Transferido'):
                motivo = str(input('\nInsira o motivo: '))
            Lista_de_ID_Doentes.remove(d)
            print("Doente removido com sucesso!")
        else:
            c-=1
        if c==0:
            print("Erro - doente não existe. Crie primeiro uma entrada…")

    md = str(input('Deseja remover outro doente? (s/n): '))
    while (md != 's') and (md != 'n'):
        md = str(input('Deseja remover outro doente? (s/n): '))
    if md == 's':
        RemoverDoente()
    else:
        return()
###########################################################################################################################
def TestarDoente():
    percentagem=0
    hoje = datetime.datetime.now()
    print('\n----------------TESTAR DOENTE----------------')
   
    nome = str(input('\nInsira o nome do Doente que pretende testar: '))
    c=len(Lista_de_ID_Doentes)
    for d in Lista_de_ID_Doentes:
        if d['nome'] == nome:
            d['data_nasc']+=' 00:00'
            data = datetime.datetime.strptime(d['data_nasc'], '%d-%m-%Y %H:%M').date()
            if data.month <= hoje.month: 
                if data.day <= hoje.day:
                    idade = hoje.year - data.year
                else:
                    idade = hoje.year - data.year - 1
            elif data.month > hoje.month: 
                idade = hoje.year - data.year - 1

            if idade >= 0 and idade <= 19:
                percentagem+= 1.25
            elif idade >= 20 and idade <= 39:    
                percentagem+= 7
            elif idade >= 40 and idade <= 59:    
                percentagem+= 9
            elif idade >= 60 and idade <= 79:    
                percentagem= 5.5
            elif idade <=80:
                percentagem+=2.25              
                
            if d['tosse'] == 's':
                percentagem += 22.5
            if (d['temp']) == 's':
                percentagem += 15.75
            if (d['dores_m']) == 's':
                percentagem += 11.25
            if (d['dores_c']) == 's':
                percentagem += 10.5
            if (d['fraqueza'])== 's':
                percentagem += 8.25
            if (d['dif_resp']) == 's':
                percentagem += 6.75
            (d['percentagem']) = percentagem

            if (d['percentagem']) >= 0 and d['percentagem']<= 30:
                (d['risco']) = 'Sem Risco'
            elif (d['percentagem']) > 30 and (d['percentagem']) <= 50:
                (d['risco']) = 'Baixo Risco'
            elif (d['percentagem']) > 50 and (d['percentagem']) <= 75:
                (d['risco']) = 'Médio Risco'
            elif (d['percentagem']) > 75:
                (d['risco']) = 'Alto Risco'

            print('\nO teste concluiu que existe {}% de probabilidade de covid-19.'.format(d['percentagem']))
            print('\nO estado do doente foi alterado para:', (d['risco']))
        else:
            c-=1
        if c==0:
            print("Erro - doente não existe. Crie primeiro uma entrada…")

    nt = str(input('\nTestar outro doente? (s/n): '))
    while (nt != 's') and (nt != 'n'):
        nt = str(input('Testar outro doente? (s/n): '))
    if nt == 's':
        TestarDoente()
    else:
        return()
###########################################################################################################################
def ConsultarDoentes():
    print('\n------------------OPÇÕES DE CONSULTA------------------')
    print('\n1 – Consultar sistema por doente')
    print('\n2 – Consultar sistema por estado')
    print('\n3 – Ver todos os doentes')
    print('\n0 – Voltar')
    return ()
###########################################################################################################################
def ConsultarPorDoente():
    print('\n------------------CONSULTAR POR DOENTE------------------')
    nome = str(input('\nInsira o nome do Doente que pretende consultar: '))

    c=len(Lista_de_ID_Doentes)
    for d in Lista_de_ID_Doentes:
        if d['nome']== nome:
            print('\nNome:', d['nome'])
            print('Tel:', d['numero_telefone'])
            print('Morada:', d['morada'])
            print('CC:', d['cc'])
            print('Data Nasc:', d['data_nasc'])
            print('Sintomas:')
            if d['tosse'] == 's':
                print('-Tosse: Sim')
            else:
                print('-Tosse: Não')
            if d['temp'] == 's':
                print('-Febre: Sim')
            else:
                print('-Febre: Não')
            if d['dores_m'] == 's':
                print('-Dores: Sim')
            else:
                print('-Dores: Não')
            if d['dores_c'] == 's':
                print(' - Cefaleias: Sim')
            else:
                print('-Cefaleias: Não')
            if d['fraqueza'] == 's':
                print('-Fraqueza: Sim')
            else:
                print('-Fraqueza: Não')
            if d['dif_resp'] == 's':
                print('-Dif. respiratória: Sim')
            else:
                print('-Dif. respiratória: Não')
            print('Probabilidade Final:', (d['percentagem']))
            print('Estado do doente:', (d['risco']))   
        else:
            c-=1
        if c==0:
            print("Erro - doente não existe. Crie primeiro uma entrada…") 
    nc = str(input('\nPretende consultar outro doente? (s/n): '))
    while (nc != 's') and (nc != 'n'):
        nc = str(input('Pretende consultar outro doente? (s/n): '))
    if nc == 's':
        ConsultarPorDoente()
    else: 
        return()
###########################################################################################################################  
def ConsultarPorEstado():
    print('\n------------------CONSULTAR POR ESTADO------------------')
    print("\nIndique qual o estado que pretende consultar")
    estado = str(input('Por Avaliar, Sem Risco, Baixo Risco, Medio Risco ou Alto Risco: '))
    while estado != 'Por Avaliar' and  estado != 'Sem Risco' and estado != 'Baixo Risco' and estado != 'Medio Risco' and  estado != 'Alto Risco':
        estado =str(input('O estado só pode ser um dos estados acima ! '))
    c=len(Lista_de_ID_Doentes)
    for d in Lista_de_ID_Doentes:
        if d['risco'] == estado:
            print('\nNome:', d['nome'])
            print('Tel:', d['numero_telefone'])
            print('Morada:', d['morada'])
            print('CC:', d['cc'])
            print('Data Nasc:', d['data_nasc'])
            print('Sintomas:')
            if d['tosse'] == 's':
                print('-Tosse: Sim')
            else:
                print('-Tosse: Não')
            if d['temp'] == 's':
                print('-Febre: Sim')
            else:
                print('-Febre: Não')
            if d['dores_m'] == 's':
                print('-Dores: Sim')
            else:
                print('-Dores: Não')
            if d['dores_c'] == 's':
                print(' - Cefaleias: Sim')
            else:
                print('-Cefaleias: Não')
            if d['fraqueza'] == 's':
                print('-Fraqueza: Sim')
            else:
                print('-Fraqueza: Não')
            if d['dif_resp'] == 's':
                print('-Dif. respiratória: Sim')
            else:
                print('-Dif. respiratória: Não')
            print('Probabilidade Final:', (d['percentagem']))
            print('Estado do doente:', (d['risco']))   
        else:
            c-=1
        if c==0:
            print("Não existe doentes com esse estado.") 
    nc = str(input('\nPretende consultar outro doente? (s/n): '))
    while (nc != 's') and (nc != 'n'):
        nc = str(input('Pretende consultar outro doente? (s/n): '))
    if nc == 's':
        ConsultarPorEstado()
    else: return() 
###########################################################################################################################
def VerTodos():
    d=len(Lista_de_ID_Doentes)
    for d in Lista_de_ID_Doentes:
        print('\nNome:', d['nome'])
        print('Tel:', d['numero_telefone'])
        print('Morada:', d['morada'])
        print('CC:', d['cc'])
        print('Data Nasc:', d['data_nasc'])
        print('Sintomas:')
        if d['tosse'] == 's':
            print('-Tosse: Sim')
        else:
            print('-Tosse: Não')
        if d['temp'] == 's':
            print('-Febre: Sim')
        else:
            print('-Febre: Não')
        if d['dores_m'] == 's':
            print('-Dores: Sim')
        else:
            print('-Dores: Não')
        if d['dores_c'] == 's':
            print(' - Cefaleias: Sim')
        else:
            print('-Cefaleias: Não')
        if d['fraqueza'] == 's':
            print('-Fraqueza: Sim')
        else:
            print('-Fraqueza: Não')
        if d['dif_resp'] == 's':
            print('-Dif. respiratória: Sim')
        else:
            print('-Dif. respiratória: Não')
        print('Probabilidade Final:', (d['percentagem']))
        print('Estado do doente:', (d['risco']))   
    if d == 0:
        print("\nNão existem dados a apresentar")
    v = str(input('\nPretende voltar? (s/n): '))
    while (v != 's') and (v != 'n'):
        v = str(input("\nPretende voltar? (s/n): "))
    if v == 'n':
        VerTodos()
    else: 
        return() 
###########################################################################################################################
def GuardarDados():
    print('\n--------GUARDAR DADOS--------')
    print('\nDeseja mesmo guardar os seus dados? (s/n): ')
    v = str(input('Atenção esta operação irá fazer overwrite a anteriores gravações! '))
    while v != 's' and v != 'n':
        v = str(input('Deseja mesmo guardar os seus dados? (s/n): '))
    if v== 's':
        with open('dados.json', 'w') as dados:
            json.dump(Lista_de_ID_Doentes, dados)
            print('Dados guardados com sucesso!')
    else:
        print('Operação cancelada, os dados não foram guardados!')
    return()
###########################################################################################################################
def CarregarDados():
    print('\n--------CARREGAR DADOS--------')
    x=str(input("Deseja carregar os ultimos dados guardados? (s/n): "))
    while x != 's' and x != 'n':
        x = str(input('Deseja carregar os ultimos dados guardados? (s/n): '))
    if x =="s":    
        dados= json.load(open('dados.json','r'))
        for a in dados:
            Lista_de_ID_Doentes.append(a)
        print("\nDados carregados com sucesso!")
    else: 
        print("Operação cancelada, os novos dados não foram carregados")
    return()
###########################################################################################################################
def Simulação():
    ocupação = modsim.State(doente=0)
    resultados = modsim.TimeSeries()

    def maisOcupação(): 
        ocupação.doente += 1
        return()

    def menosOcupação():
        if ocupação.doente >= 0:
            ocupação.doente -= 1
        return()

    def Dias(x,y):
        if (x < y/4):
            if modsim.flip(0.3):
                maisOcupação()
            if modsim.flip(0.1):
                menosOcupação()
        if(x < y/2):
            if modsim.flip(0.6):
                maisOcupação()
            if modsim.flip(0.3):
                menosOcupação()
        if(x >(3/4)*y):
            if modsim.flip(0.2):
                maisOcupação()
            if modsim.flip(0.5):
                menosOcupação()
        if(x >(3/4)*y):
            if modsim.flip(0.1):
                maisOcupação()
            if modsim.flip(0.7):
                menosOcupação()
        return()
    def RunSimulation(time):
        for a in range(time) :
            Dias(a,time) 
            resultados[a] = ocupação.doente 
        return()
    def Gráfico():
        modsim.plot(resultados, label="UCI")
        modsim.decorate(title="Doentes Internados",xlabel="Tempo",ylabel="Número de doentes Internados")
        modsim.plt.show()
        return()
    
    dias = str(input("Digite o número de dias da simulação: "))
    while not dias.isdigit() or int(dias)<4:        
        dias = str(input("Digite o número de dias da simulação (mínimo 4): "))
    Simulacao = RunSimulation(int(dias))
    Gráfico()
    v = str(input('\nPretende correr uma nova simulação? (s/n): '))
    while (v != 's') and (v != 'n'):
        v = str(input("\nPretende correr uma nova simulação? (s/n): "))
    if v == 'n':
        return()
    else: Simulação()
###########################################################################################################################
def UI():
    es = 'Cthulhu'
    while es != '0':
        if es == '1':
            AdicionarDoente()      
        elif es == '2':
            ModificarSintomas()
        elif es == '3':
            TestarDoente()
        elif es == '4':
            RemoverDoente()
        elif es == '5':     
            esc = "Azathoth"
            while esc != '0':
                if esc == '1':
                    ConsultarPorDoente()
                elif esc == '2':
                    ConsultarPorEstado()
                elif esc == '3':
                    VerTodos()
                elif esc == '0':
                    MenuPrincipal()
                ConsultarDoentes()
                esc = str(input('\nDigite a sua operação: '))
        elif es == '6':
            GuardarDados()
        elif es == '7':
            CarregarDados()
        elif es == '8':
            Simulação()
        MenuPrincipal()
        es = str(input('\nEscolha uma opção: '))
    if es == '0':
        sair = str(input('Os seus dados não guardados não serão salvos, deseja mesmo sair? (s/n): '))
        while (sair != 's') and (sair != 'n'):
            sair = str(input('Os seus dados não guardados não serão salvos, deseja mesmo sair? (s/n): '))
        if sair == "s":
            print('Fim!!')
        else:
            UI()
###########################################################################################################################            
UI()
```
